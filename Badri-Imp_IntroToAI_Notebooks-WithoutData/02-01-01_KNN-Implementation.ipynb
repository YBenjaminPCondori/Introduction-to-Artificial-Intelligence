# Imports

```python
import numpy as np
import pandas as pd
import math

import os

import matplotlib.pyplot as plt

import sklearn

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
```

# Getting File Path

```python
path = os.getcwd()
files = os.listdir(path)

newPath = path+'\\mergedCleanedCRDataset.csv'
print(newPath)
```

```python
crDf = pd.read_csv(newPath,sep='\t')
crDf.head()
```

# Label Encoding

Converting categorical columns into numerical ones

```python
crEncDf = crDf.copy()

for i in crEncDf.columns:
    if not (i == 'Longitude' or i=='Latitude'):
        labEnc = LabelEncoder()
        crEncDf[i] = labEnc.fit_transform(crEncDf[i])

crEncDf.head()
```

# Location Data Normalisation

Here we are normalising the Longitude and Latitude between a 0 to 1 scale, we may lose some data, but it ensures that it is easier for the models to process those columns

```python
crScalDf = crEncDf.copy()

for i in crScalDf.columns:
    if (i == 'Longitude' or i=='Latitude'):
        minMaxScal = MinMaxScaler()
        crScalDf[i] = minMaxScal.fit_transform(crScalDf[i].to_numpy().reshape(-1,1)).flatten()

crScalDf.head()
        
```

# Train Test Split

Train Size is 70%
Test Size is 30% 
Random State is 16

```python
x = crScalDf.drop(columns='Crime type')
y = crScalDf['Crime type']

xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size=0.3,shuffle=True,random_state=16)
```

# Running K- Nearest Neighbour Algorithm

Implemented a primitive batch fitting

```python
knnBase1 = KNeighborsClassifier()
batchSize = 10000
maxSize = xTrain.size
for i in range(0,maxSize-batchSize,batchSize):
    knnBase1.fit(xTrain[i:i+batchSize],yTrain[i:i+batchSize])
    print(i)

```

```python
accListX = []
accListY = []

yPred1 = knnBase1.predict(xTest)
acc1 = accuracy_score(yTest,yPred1)
#accListX.append(i)
accListY.append(acc1)


#plt.plot(accListY)
print(np.max(accListY))
```

# Running K- Nearest Neighbour Algorithm with the number of neighbours equal to the number of classes in target column

Implemented a primitive batch fitting

```python
knnBase2 = KNeighborsClassifier(n_neighbors=14)
batchSize = 10000
maxSize = xTrain.size
for i in range(0,maxSize,batchSize):
    knnBase2.fit(xTrain[i:i+batchSize],yTrain[i:i+batchSize])
    print(i)

```

```python
accListX = []
accListY = []

yPred2 = knnBase2.predict(xTest)
acc2 = accuracy_score(yTest,yPred2)
#accListX.append(i)
accListY.append(acc2)


#plt.plot(accListY)
print(np.max(accListY))
```
