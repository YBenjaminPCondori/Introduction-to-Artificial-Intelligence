# Importing multiple excel files into pandas dataframe

Source: https://stackoverflow.com/questions/20908018/import-multiple-excel-files-into-python-pandas-and-concatenate-them-into-one-dat

```python
import os
import pandas as pd

path = os.getcwd()
path = path+'/Dataset'
files = os.listdir(path)

print(files)
```

## Navigating via the files and converting each file into pandas dataframe and then appending them into a singular dataframe

```python
crDf = pd.DataFrame()

for f in files:
    newPath = path+'/'+f
    newFiles = os.listdir(newPath)
    for nf in newFiles:
        data = pd.read_csv(newPath+'/'+nf)
        crDf = crDf.append(data)
        print(nf+' was completed')
```
